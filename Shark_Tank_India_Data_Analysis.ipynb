```python
selected_columns = [
    'Season Number',
    'Startup Name',
    'Industry',
    'Pitchers City',
    'Pitchers State',
    'Yearly Revenue',
    'Monthly Sales',
    'Original Ask Amount',
    'Original Offered Equity',
    'Valuation Requested',
    'Received Offer',
    'Accepted Offer',
    'Total Deal Amount',
    'Total Deal Equity',
    'Deal Valuation',
    'Number of Sharks in Deal'
]

project_df = df[selected_columns]

project_df.head()
```

```python
project_df.isnull().sum().sort_values(ascending=False)
```

```python
project_df['Accepted Offer'].value_counts(dropna=False)
```

```python
pd.crosstab(
    project_df['Received Offer'],
    project_df['Accepted Offer'],
    dropna=False
)
```

```python
project_df = project_df.copy()

project_df['Accepted Offer'] = project_df['Accepted Offer'].fillna(0)

project_df['Pitchers City'] = project_df['Pitchers City'].fillna('Unknown')
project_df['Pitchers State'] = project_df['Pitchers State'].fillna('Unknown')

project_df.isnull().sum().sort_values(ascending=False)
```

```python
project_df.to_csv("shark_tank_cleaned.csv", index=False)
```

```python
project_df.shape
```

```python
total_pitches = len(project_df)
offers_received = project_df['Received Offer'].sum()
deals_accepted = project_df['Accepted Offer'].sum()
total_investment = project_df['Total Deal Amount'].sum()

print("Total Pitches:", total_pitches)
print("Offers Received:", int(offers_received))
print("Deals Accepted:", int(deals_accepted))
print("Total Investment:", total_investment)
```

```python
offer_rate = (offers_received / total_pitches) * 100
deal_success_rate = (deals_accepted / total_pitches) * 100
offer_acceptance_rate = (deals_accepted / offers_received) * 100

print("Offer Rate:", round(offer_rate, 2), "%")
print("Overall Deal Success Rate:", round(deal_success_rate, 2), "%")
print("Offer Acceptance Rate:", round(offer_acceptance_rate, 2), "%")
```

```python
industry_investment = (
    project_df.groupby('Industry')['Total Deal Amount']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

industry_investment
```

```python
industry_pitches = (
    project_df['Industry']
    .value_counts()
    .head(10)
)

industry_pitches
```

```python
industry_success = project_df.groupby('Industry').agg(
    Total_Pitches=('Startup Name', 'count'),
    Deals_Accepted=('Accepted Offer', 'sum')
)

industry_success['Success_Rate'] = (
    industry_success['Deals_Accepted'] /
    industry_success['Total_Pitches'] * 100
)

industry_success[
    industry_success['Total_Pitches'] >= 20
].sort_values('Success_Rate', ascending=False)
```

```python
season_analysis = project_df.groupby('Season Number').agg(
    Total_Pitches=('Startup Name', 'count'),
    Deals_Accepted=('Accepted Offer', 'sum'),
    Total_Investment=('Total Deal Amount', 'sum')
)

season_analysis['Deal_Success_Rate'] = (
    season_analysis['Deals_Accepted'] /
    season_analysis['Total_Pitches'] * 100
)

season_analysis
```

```python
season_analysis['Average_Deal_Size'] = (
    season_analysis['Total_Investment'] /
    season_analysis['Deals_Accepted']
)

season_analysis.round(2)
```

```python
top_deals = project_df[
    project_df['Accepted Offer'] == 1
][
    ['Startup Name', 'Industry', 'Season Number', 'Total Deal Amount', 'Deal Valuation']
].sort_values(
    'Total Deal Amount',
    ascending=False
).head(10)

top_deals
```

```python
valuation_analysis = project_df[
    project_df['Accepted Offer'] == 1
][
    ['Startup Name', 'Valuation Requested', 'Deal Valuation']
].dropna().copy()

valuation_analysis['Valuation Change %'] = (
    (valuation_analysis['Deal Valuation'] - valuation_analysis['Valuation Requested'])
    / valuation_analysis['Valuation Requested'] * 100
)

valuation_analysis['Valuation Change %'].describe()
```

```python
valuation_analysis.sort_values(
    'Valuation Change %',
    ascending=False
).head(10)
```

```python
valuation_filtered = valuation_analysis[
    valuation_analysis['Valuation Requested'] >= 1
]

valuation_filtered['Valuation Change %'].describe()
```

```python
industry_investment.sort_values().plot(
    kind='barh',
    figsize=(10, 6)
)

plt.title('Top 10 Industries by Total Investment')
plt.xlabel('Total Deal Amount')
plt.ylabel('Industry')
plt.tight_layout()

plt.savefig("industry_investment.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
season_analysis['Total_Investment'].plot(
    kind='line',
    marker='o',
    figsize=(8, 5)
)

plt.title('Total Investment by Season')
plt.xlabel('Season Number')
plt.ylabel('Total Deal Amount')
plt.xticks(season_analysis.index)
plt.grid(True)
plt.tight_layout()

plt.savefig("season_investment.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
industry_success_filtered = industry_success[
    industry_success['Total_Pitches'] >= 20
].sort_values('Success_Rate')

industry_success_filtered['Success_Rate'].plot(
    kind='barh',
    figsize=(10, 6)
)

plt.title('Deal Success Rate by Industry')
plt.xlabel('Success Rate (%)')
plt.ylabel('Industry')
plt.tight_layout()

plt.savefig("industry_success_rate.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
season_analysis['Average_Deal_Size'].plot(
    kind='bar',
    figsize=(8, 5)
)

plt.title('Average Deal Size by Season')
plt.xlabel('Season Number')
plt.ylabel('Average Deal Amount')
plt.xticks(rotation=0)
plt.tight_layout()

plt.savefig("average_deal_size.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
top_deals_plot = top_deals.sort_values('Total Deal Amount')

plt.figure(figsize=(10, 6))

plt.barh(
    top_deals_plot['Startup Name'],
    top_deals_plot['Total Deal Amount']
)

plt.title('Top 10 Biggest Deals in Shark Tank India')
plt.xlabel('Total Deal Amount')
plt.ylabel('Startup Name')
plt.tight_layout()

plt.savefig("top_10_deals.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(
    valuation_filtered['Valuation Requested'],
    valuation_filtered['Deal Valuation'],
    alpha=0.6
)

plt.title('Requested Valuation vs Deal Valuation')
plt.xlabel('Valuation Requested')
plt.ylabel('Deal Valuation')
plt.tight_layout()

plt.savefig("valuation_comparison.png", dpi=300, bbox_inches="tight")
plt.show()
```
